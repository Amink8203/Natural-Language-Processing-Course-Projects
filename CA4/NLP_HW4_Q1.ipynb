```python
class BARTDataset(Dataset):
    def __init__(self, df, tokenizer, max_inp_tok=1024, max_out_tok=256):
        self.data = df.reset_index(drop=True)
        self.tokenizer = tokenizer
        self.max_inp_tok = max_inp_tok
        self.max_out_tok = max_out_tok

    def __len__(self):
        return len(self.data)

    def __getitem__(self, index):
        row = self.data.iloc[index]

        inp_text = (
            "Task: Translate given natural language Question about given Schema into SQL.\n"
            f"Question: {row['question']}\n"
            f"Schema: {row['schema']}"
        )
        target_text = row['query']

        inp_encoding = self.tokenizer(
            inp_text,
            max_length=self.max_inp_tok,
            padding='max_length',
            truncation=True,
            return_tensors='pt'
        )

        target_encoding = self.tokenizer(
            target_text,
            max_length=self.max_out_tok,
            padding='max_length',
            truncation=True,
            return_tensors='pt'
        )

        labels = target_encoding['input_ids'].squeeze()
        # if it was a pad token, we should mask it (mask value for torch is -100)
        # in order to not consider these tokens in learning process
        labels[labels == self.tokenizer.pad_token_id] = -100

        return {
            'input_ids': inp_encoding['input_ids'].squeeze(),
            'attention_mask': inp_encoding['attention_mask'].squeeze(),
            'labels': labels
        }
```

```python
# function for generating sql with trained model
def generate_sql(model, tokenizer, question, schema, device, max_len=256):
    model.eval()
    inp_text = (
            "Task: Translate given natural language Question about given Schema into SQL.\n"
            f"Question: {question}\n"
            f"Schema: {schema}"
        )

    inputs = tokenizer(
        inp_text,
        max_length=1024,
        truncation=True,
        return_tensors='pt'
    ).to(device)

    with torch.no_grad():
        output_ids = model.generate(
            input_ids=inputs['input_ids'],
            attention_mask=inputs['attention_mask'],
            max_length=max_len,
            num_beams=4,
            early_stopping=True
        )

    sql = tokenizer.decode(output_ids[0], skip_special_tokens=True)
    return sql
```

```python
def raw_exact_match(pred, actual):
    return pred.strip() == actual.strip()
```

```python
def normalize_sql(sql):
    sql = sql.lower().strip()
    sql = sql.replace(';', '')

    # normalize operators spacing
    sql = re.sub(r'\s*=\s*', ' = ', sql)
    sql = re.sub(r'\s*<\s*', ' < ', sql)
    sql = re.sub(r'\s*>\s*', ' > ', sql)
    sql = re.sub(r'\s*<=\s*', ' <= ', sql)
    sql = re.sub(r'\s*>=\s*', ' >= ', sql)
    sql = re.sub(r'\s*!=\s*', ' != ', sql)
    sql = sqlparse.format(
        sql,
        reindent=False,
        keyword_case='lower', # lower the sql keywords
        strip_comments=True
    )
    sql = ' '.join(sql.split())
    return sql

def normalized_exact_match(pred, actual):
    pred_norm = normalize_sql(pred)
    actual_norm = normalize_sql(actual)
    return pred_norm == actual_norm
```

```python
print("Testing normalization:")
test_sql1 = "SELECT  sname FrOM   students WHERE sid = 1;"
test_sql2 = "select snaMe fRom students where sid = 1"
print(f"SQL 1: '{test_sql1}'")
print(f"SQL 2: '{test_sql2}'")
print(f"Normalized 1: '{normalize_sql(test_sql1)}'")
print(f"Normalized 2: '{normalize_sql(test_sql2)}'")
print(f"Raw match: {raw_exact_match(test_sql1, test_sql2)}")
print(f"Normalized match: {normalized_exact_match(test_sql1, test_sql2)}")
```

```python
def evaluate_model(model, tokenizer, df, device, split_name):
    raw_correct = 0
    norm_correct = 0
    total = len(df)
    results = []

    print(f"Evaluating on {split_name} set ({total} samples)\n")

    for idx in tqdm(range(total), desc=split_name):
        row = df.iloc[idx]
        pred = generate_sql(model, tokenizer, row['question'], row['schema'], device)
        gold = row['query']

        is_raw = raw_exact_match(pred, gold)
        is_norm = normalized_exact_match(pred, gold)

        if is_raw:
            raw_correct += 1
        if is_norm:
            norm_correct += 1

        results.append({
            'question': row['question'],
            'schema': row['schema'],
            'predicted': pred,
            'gold': gold,
            'raw_match': is_raw,
            'norm_match': is_norm
        })

    raw_em = (raw_correct / total) * 100
    norm_em = (norm_correct / total) * 100

    return raw_em, norm_em, results
```

```python
class GPT2Dataset(Dataset):
    def __init__(self, dataframe, tokenizer, max_length):
        self.data = dataframe.reset_index(drop=True)
        self.tokenizer = tokenizer
        self.max_length = max_length

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        row = self.data.iloc[idx]
        question = str(row["question"])
        schema = str(row["schema"])
        query = str(row["query"])

        prefix, full_sequence = format_gpt2_input(question, schema, query)

        prefix_tokens = self.tokenizer(prefix, add_special_tokens=False)
        prefix_length = len(prefix_tokens["input_ids"])

        encoding = self.tokenizer(
            full_sequence,
            max_length=self.max_length - 1,
            truncation=True,
            padding=False,
            add_special_tokens=False
        )

        input_ids = encoding["input_ids"] + [self.tokenizer.eos_token_id]
        attention_mask = [1] * len(input_ids)

        labels = input_ids.copy()
        # mask prefix tokens
        for i in range(min(prefix_length, len(labels))):
            labels[i] = -100

        padding_length = self.max_length - len(input_ids)
        # apply padding to sequences that are less than max_length
        if padding_length > 0:
            input_ids = input_ids + [self.tokenizer.pad_token_id] * padding_length
            attention_mask = attention_mask + [0] * padding_length
            labels = labels + [-100] * padding_length

        return {
            "input_ids": torch.tensor(input_ids, dtype=torch.long),
            "attention_mask": torch.tensor(attention_mask, dtype=torch.long),
            "labels": torch.tensor(labels, dtype=torch.long),
            "prefix_length": prefix_length
        }
```

```python
def generate_sql_gpt2(model, tokenizer, question, schema, device, max_new_tokens=150):
    model.eval()

    prefix, _ = format_gpt2_input(question, schema)

    inputs = tokenizer(prefix, return_tensors="pt", add_special_tokens=False)
    input_ids = inputs["input_ids"].to(device)
    attention_mask = inputs["attention_mask"].to(device)

    prefix_length = input_ids.shape[1]

    with torch.no_grad():
        outputs = model.generate(
            input_ids=input_ids,
            attention_mask=attention_mask,
            max_new_tokens=max_new_tokens,
            num_beams=4,
            early_stopping=True,
            pad_token_id=tokenizer.pad_token_id,
            eos_token_id=tokenizer.eos_token_id,
            do_sample=False
        )

    generated_ids = outputs[0][prefix_length:]
    sql_output = tokenizer.decode(generated_ids, skip_special_tokens=True).strip()

    # handle potential repetition or garbage at the end
    lines = sql_output.split('\n')
    sql_output = lines[0].strip()

    return sql_output
```

```python
def evaluate_model_gpt2(model, tokenizer, df, device, split_name):
    raw_correct = 0
    norm_correct = 0
    total = len(df)
    results = []

    print(f"Evaluating on {split_name} set ({total} samples)\n")

    for idx in tqdm(range(total), desc=split_name):
        row = df.iloc[idx]
        pred = generate_sql_gpt2(model, tokenizer, row['question'], row['schema'], device)
        gold = row['query']

        is_raw = raw_exact_match(pred, gold)
        is_norm = normalized_exact_match(pred, gold)

        if is_raw:
            raw_correct += 1
        if is_norm:
            norm_correct += 1

        results.append({
            'question': row['question'],
            'schema': row['schema'],
            'predicted': pred,
            'gold': gold,
            'raw_match': is_raw,
            'norm_match': is_norm
        })

    raw_em = (raw_correct / total) * 100
    norm_em = (norm_correct / total) * 100

    return raw_em, norm_em, results
```
