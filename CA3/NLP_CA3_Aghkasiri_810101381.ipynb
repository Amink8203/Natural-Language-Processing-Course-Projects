```python
UNK_TOKEN = "<UNK>"
UNK_INDEX = 0
```

```python
def preprocess_text(txt):
    txt = txt.lower()
    txt = re.sub(r'[^a-z0-9\s]', '', txt)
    return txt.split()
```

```python
def build_vocab(wordToidx, idxToword, dataset, min_freq):
    counter = Counter()
    for line in dataset:
        tokens = preprocess_text(line)
        counter.update(tokens)
    wordToidx = {UNK_TOKEN: UNK_INDEX}
    idxToword = {UNK_INDEX: UNK_TOKEN}
    i = 1
    for word, freq in counter.items():
        if freq >= min_freq:
            wordToidx[word] = i
            idxToword[i] = word
            i += 1
    return wordToidx, idxToword
```

```python
def change_text_to_nums(wordToidx, text):
    tokens = preprocess_text(text)
    return [wordToidx.get(token, UNK_INDEX) for token in tokens]
```

```python
class WikiDataset(Dataset):
    def __init__(self, texts_list, wordToidx):
        self.data = []
        for text in texts_list:
            nums = change_text_to_nums(wordToidx, text)
            if len(nums) > 0:
                self.data.append(nums)
    def __len__(self):
        return len(self.data)
    def __getitem__(self, idx):
        return self.data[idx]
```

```python
def cbow_collate_fn(batch, window_size):
    inputs = []
    targets = []
    for sentence in batch:
        if len(sentence) < 2 * window_size + 1:
            continue
        for i in range(window_size, len(sentence) - window_size):
            context = sentence[i - window_size : i] + sentence[i + 1 : i + window_size + 1]
            target = sentence[i]
            inputs.append(context)
            targets.append(target)
    if not inputs:
        return torch.tensor([]), torch.tensor([])
    return torch.tensor(inputs), torch.tensor(targets)
```

```python
def skipgram_collate_fn(batch, window_size):
    inputs = []
    targets = []
    for sentence in batch:
        if len(sentence) < 2 * window_size + 1:
            continue
        for i in range(window_size, len(sentence) - window_size):
            center = sentence[i]
            context_indices = list(range(i - window_size, i)) + list(range(i + 1, i + window_size + 1))
            for j in context_indices:
                inputs.append(center)
                targets.append(sentence[j])
    if not inputs:
        return torch.tensor([]), torch.tensor([])
    return torch.tensor(inputs), torch.tensor(targets)
```

```python
# Predict the center word based on the sum of context words
class CBOW(nn.Module):
    def __init__(self, vocab_size, embed_size):
        super(CBOW, self).__init__()
        # Input (context) embeddings
        self.input_emb = nn.Embedding(vocab_size, embed_size)
        # Output (prediction) embeddings
        self.output_emb = nn.Embedding(vocab_size, embed_size)
        # Initialization (small uniform values)
        init_range = 0.5 / embed_size
        nn.init.uniform_(self.input_emb.weight, -init_range, init_range)
        nn.init.uniform_(self.output_emb.weight, -init_range, init_range)

    def forward(self, inputs):
      # Look up input (context) embeddings
      embeds = self.input_emb(inputs)
      # Combine context vectors by averaging them
      embeds_mean = embeds.mean(dim=1)
      # Compute logits by dotting with output embedding matrix
      logits = torch.matmul(embeds_mean, self.output_emb.weight.t())
      return logits
```

```python
# Predict a context word based on the center word (I have not used negative sampling)
class SkipGram(nn.Module):
    def __init__(self, vocab_size, embed_size):
        super(SkipGram, self).__init__()
        # Center embeddings
        self.input_emb = nn.Embedding(vocab_size, embed_size)
        # Context(output) embeddings
        self.output_emb = nn.Embedding(vocab_size, embed_size)
        init_range = 0.5 / embed_size
        nn.init.uniform_(self.input_emb.weight, -init_range, init_range)
        nn.init.uniform_(self.output_emb.weight, -init_range, init_range)

    def forward(self, inputs):
        embed = self.input_emb(inputs)
        logits = torch.matmul(embed, self.output_emb.weight.t())
        return logits

# Note: I have not put softmax in above 2 classes because the outputs of linear are logits
# and when they are passed to CrossEntropyLoss, it automatically applies Softmax internally before calculating the error
```

```python
def train_epoch(model, dataloader, optimizer, criterion, device, desc="Train"):
    model.train()
    total_loss = 0.0
    total_tokens = 0
    progress = tqdm(dataloader, desc=desc, leave=True)
    for batch in progress:
        inputs, targets = batch
        if inputs.numel() == 0:
            continue
        inputs = inputs.to(device)
        targets = targets.to(device)
        optimizer.zero_grad()
        logits = model(inputs)
        loss = criterion(logits, targets)
        loss.backward()
        optimizer.step()
        batch_tokens = targets.numel()
        total_loss += loss.item() * batch_tokens
        total_tokens += batch_tokens
        avg = (total_loss / total_tokens) if total_tokens > 0 else 0.0
        progress.set_postfix(batch_loss=f"{loss.item():.4f}", avg_loss=f"{avg:.4f}")
    if total_tokens == 0:
        return None
    return total_loss / total_tokens

def eval_epoch(model, dataloader, criterion, device, desc="Eval"):
    model.eval()
    total_loss = 0.0
    total_tokens = 0
    progress = tqdm(dataloader, desc=desc, leave=True)
    with torch.no_grad():
        for batch in progress:
            inputs, targets = batch
            if inputs.numel() == 0:
                continue
            inputs = inputs.to(device)
            targets = targets.to(device)
            logits = model(inputs)
            loss = criterion(logits, targets)
            batch_tokens = targets.numel()
            total_loss += loss.item() * batch_tokens
            total_tokens += batch_tokens
            avg = (total_loss / total_tokens) if total_tokens > 0 else 0.0
            progress.set_postfix(batch_loss=f"{loss.item():.4f}", avg_loss=f"{avg:.4f}")
    if total_tokens == 0:
        return None
    return total_loss / total_tokens

def run(train_loader, valid_loader, test_loader, mode="cbow", embed_size=150, window_size=4, batch_size=32, lr=0.001, epochs=100, device=None):
    if mode == "cbow":
        model = CBOW(vocab_size, embed_size).to(device)
        collate = partial(cbow_collate_fn, window_size=window_size)
    else:
        model = SkipGram(vocab_size, embed_size).to(device)
        collate = partial(skipgram_collate_fn, window_size=window_size)

    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=lr)

    train_losses = []
    valid_losses = []
    test_losses  = []

    for epoch in range(1, epochs + 1):
        desc_train = f"{mode.upper()} Epoch {epoch}/{epochs} Train"
        desc_valid = f"{mode.upper()} Epoch {epoch}/{epochs} Valid"
        desc_test  = f"{mode.upper()} Epoch {epoch}/{epochs} Test"

        train_loss = train_epoch(model, train_loader, optimizer, criterion, device, desc=desc_train)
        valid_loss = eval_epoch(model, valid_loader, criterion, device, desc=desc_valid)
        test_loss  = eval_epoch(model, test_loader,  criterion, device, desc=desc_test)

        train_losses.append(train_loss)
        valid_losses.append(valid_loss)
        test_losses.append(test_loss)

    print("\nFinal Test Loss =", test_losses[-1])
    return model, train_losses, valid_losses, test_losses
```

```python
def plot_losses(train_losses, valid_losses, test_losses, mode="CBOW"):
    epochs = np.arange(1, len(train_losses) + 1)
    plt.figure(figsize=(8,5))
    plt.plot(epochs, train_losses, label="Train Loss", color="blue")
    plt.plot(epochs, valid_losses, label="Validation Loss", color="red")
    plt.plot(epochs, test_losses, label="Test Loss", color="green")
    plt.title(f"{mode.upper()} Loss Curves")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
# get the embedding of the word we want from embedding layer of our model
def get_word_embedding(model, word, wordToidx, device):
    if word not in wordToidx:
        return None
    word_idx = wordToidx[word]
    word_tensor = torch.tensor([word_idx]).to(device)
    with torch.no_grad():
        embedding = model.output_emb(word_tensor).cpu().numpy()
    return embedding.reshape(-1)
```

```python
# get embeddings of all of our vocab's words (with dimenstion (vocab_size,embed_size))
def get_all_embeddings(model, wordToidx, device):
    vocab_size = len(wordToidx)
    all_indices = torch.arange(vocab_size).to(device)
    with torch.no_grad():
        embeddings = model.output_emb(all_indices).cpu().numpy()
    return embeddings
```

```python
# obtain top 5 most similar words to our target word (based on cosine similarity)
def find_similar_words(model, target_word, wordToidx, idxToword, device, words_num=5):
    target_embedding = get_word_embedding(model, target_word, wordToidx, device)
    if target_embedding is None:
        print(f"Word '{target_word}' not in vocabulary")
        return []

    all_embeddings = get_all_embeddings(model, wordToidx, device)
    target_embedding = target_embedding.reshape(1, -1)
    similarities = cosine_similarity(target_embedding, all_embeddings)[0]

    target_idx = wordToidx[target_word]
    similarities[target_idx] = -1

    top_indices = np.argsort(similarities)[::-1][:words_num]
    similar_words = [(idxToword[idx], similarities[idx]) for idx in top_indices]

    return similar_words
```

```python
# visualize with tsne
def visualize_tsne(model, target_word, wordToidx, idxToword, device, words_num=5, model_name="CBOW"):
    similar_words = find_similar_words(model, target_word, wordToidx, idxToword, device, words_num)

    if not similar_words:
        return

    words_to_plot = [target_word] + [word for word, _ in similar_words]
    embeddings_list = []

    for word in words_to_plot:
        emb = get_word_embedding(model, word, wordToidx, device)
        embeddings_list.append(emb)

    embeddings_arr = np.array(embeddings_list)

    tsne = TSNE(n_components=2, random_state=30, perplexity=min(5, len(words_to_plot)-1))
    embeddings_2d = tsne.fit_transform(embeddings_arr)

    plt.figure(figsize=(10, 8))

    plt.scatter(embeddings_2d[0, 0], embeddings_2d[0, 1], c='red', s=200, marker='*', linewidths=2, label=f'Target: {target_word}', zorder=3)

    for i in range(1, len(words_to_plot)):
        plt.scatter(embeddings_2d[i, 0], embeddings_2d[i, 1],
                   c='blue', s=100, alpha=0.6, edgecolors='black', linewidths=1)
        plt.annotate(words_to_plot[i], xy=(embeddings_2d[i, 0], embeddings_2d[i, 1]), xytext=(5, 5), textcoords='offset points', fontsize=10, alpha=0.8)

    plt.annotate(target_word, xy=(embeddings_2d[0, 0], embeddings_2d[0, 1]), xytext=(5, 5), textcoords='offset points', fontsize=12, fontweight='bold', color='red')

    plt.title(f'{model_name}: t-SNE Visualization of "{target_word}" and Similar Words', fontsize=14, fontweight='bold')
    plt.xlabel('t-SNE Dimension 1', fontsize=12)
    plt.ylabel('t-SNE Dimension 2', fontsize=12)
    plt.legend(fontsize=10)
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()

    print(f"{model_name} - Similar words to '{target_word}':")
    for word, score in similar_words:
        print(f"  {word}: {score:.4f}")
```

```python
def preprocess_text(text):
    text = text.lower()
    text = ' '.join(text.split())
    return text
```

```python
# gives equal number of samples for each class
def get_balanced_samples(dataset, samples_num, classes_num=4):
    samples_per_class = samples_num // classes_num

    # a dictionary for holding the indices of corresponding class numbers of dataset
    class_indices = {i: [] for i in range(classes_num)}
    for i, label in enumerate(dataset['label']):
        class_indices[label].append(i)

    selected_indices = []
    for i in range(classes_num):
        index = class_indices[i]
        np.random.shuffle(index)
        selected_indices.extend(index[:samples_per_class])

    np.random.shuffle(selected_indices)

    texts = [dataset['text'][i] for i in selected_indices]
    labels = [dataset['label'][i] for i in selected_indices]

    return texts, labels
```

```python
def get_sentence_embedding(text, model):
    words = text.split()
    if len(words) == 0:
        return np.zeros(model.get_dimension())

    word_vectors = []
    for word in words:
        word_vectors.append(model.get_word_vector(word))

    sentence_vector = np.mean(word_vectors, axis=0)
    return sentence_vector

def embed_texts(texts, model):
    embeddings = []
    for text in texts:
        embedding = get_sentence_embedding(text, model)
        embeddings.append(embedding)
    return np.array(embeddings)
```

```python
class EmbedDataset(Dataset):
    def __init__(self, embeddings, labels):
        self.embeddings = torch.FloatTensor(embeddings)
        self.labels = torch.LongTensor(labels)

    def __len__(self):
        return len(self.labels)

    def __getitem__(self, idx):
        return self.embeddings[idx], self.labels[idx]
```

```python
class AGNewsNN(nn.Module):
    def __init__(self, inp=300, out=4, dropout_rate=0.3):
        super(AGNewsNN, self).__init__()
        self.fc1 = nn.Linear(inp, 256)

        self.fc2 = nn.Linear(256, 128)
        self.do2 = nn.Dropout(dropout_rate)

        self.fc3 = nn.Linear(128, out)

        self.relu = nn.ReLU()

    def forward(self, x):
        x = self.fc1(x)
        x = self.relu(x)

        x = self.fc2(x)
        x = self.relu(x)
        x = self.do2(x)

        x = self.fc3(x)
        return x
```

```python
def train_epoch(model, dataloader, criterion, optimizer, device):
    model.train()
    total_loss = 0
    correct = 0
    total = 0

    pbar = tqdm(dataloader, desc='Training', leave=True)
    for embeddings, labels in pbar:
        embeddings = embeddings.to(device)
        labels = labels.to(device)

        optimizer.zero_grad()
        outputs = model(embeddings)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        total_loss += loss.item()
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

        pbar.set_postfix({'loss': f'{loss.item():.4f}', 'acc': f'{100*correct/total:.2f}%'})

    avg_loss = total_loss / len(dataloader)
    accuracy = 100 * correct / total
    return avg_loss, accuracy

def evaluate(model, dataloader, criterion, device):
    model.eval()
    total_loss = 0
    correct = 0
    total = 0

    pbar = tqdm(dataloader, desc='Evaluating', leave=True)
    with torch.no_grad():
        for embeddings, labels in pbar:
            embeddings = embeddings.to(device)
            labels = labels.to(device)

            outputs = model(embeddings)
            loss = criterion(outputs, labels)

            total_loss += loss.item()
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

            pbar.set_postfix({'loss': f'{loss.item():.4f}', 'acc': f'{100*correct/total:.2f}%'})

    avg_loss = total_loss / len(dataloader)
    accuracy = 100 * correct / total
    return avg_loss, accuracy
```
