```python
import pandas as pd
import re
from sklearn.feature_extraction.text import CountVectorizer
import numpy as np
```

```python
df = pd.read_csv('./datasets/q1/emails_1.csv')
df['status'] = df['status'].map({'ham': 0, 'spam': 1})
```

```python
df.head()
```

```python
def clean_text(text):
    text = text.lower()                          
    text = re.sub(r'[^a-z\s]', '', text)          
    text = re.sub(r'\s+', ' ', text).strip()     
    return text

df["clean_text"] = df["text"].apply(clean_text)
df.head()
```

```python
vectorizer = CountVectorizer(max_features=15)
counts = vectorizer.fit_transform(df["clean_text"]).toarray()
vocab = vectorizer.get_feature_names_out()
new_df = pd.DataFrame(counts, columns=vocab)
new_df['status'] = df['status']
new_df.head()
```

```python
import numpy as np
def get_accuracy(y_true, y_pred):
    correct_predictions = np.sum(y_true == y_pred)
    total_predictions = len(y_true)
    return correct_predictions / total_predictions
def get_precision(y_true, y_pred):
    tp = np.sum((y_true == 1) & (y_pred == 1))
    predicted_pos = np.sum(y_pred == 1)
    return tp / predicted_pos if predicted_pos > 0 else 0
def get_recall(y_true, y_pred):
    tp = np.sum((y_true == 1) & (y_pred == 1))
    actual_pos = np.sum(y_true == 1)
    return tp / actual_pos if actual_pos > 0 else 0
def f1_score(y_true, y_pred):
    precision = get_precision(y_true, y_pred)
    recall = get_recall(y_true, y_pred)
    return (2* (precision * recall)) / (precision + recall) if (precision + recall) > 0 else 0
```

```python
y_true = np.array([0,1,1,0,1])
y_pred = np.array([0,1,0,0,1])
print(f"Accuracy: {get_accuracy(y_true, y_pred):.2f}")
print(f"Precision: {get_precision(y_true, y_pred):.2f}")
print(f"Recall: {get_recall(y_true, y_pred):.2f}")
print(f"F1 Score: {f1_score(y_true, y_pred):.2f}")
```

```python
class MultinomialNaiveBayes:
    def __init__(self):
        self.class_probs = None # P(C_k)
        self.word_probs = None # P(x_i | C_k)
        self.classes = None              
    
    def fit(self, X, y):
        smaples_num, features_num = X.shape
        self.classes = np.unique(y)        
        classes_num = len(self.classes)
        self.class_probs = np.zeros(classes_num)
        self.word_probs = np.zeros((classes_num, features_num))
        
        for i in range(len(self.classes)):
            c = self.classes[i]
            X_class = X[y == c]
            self.class_probs[i] = X_class.shape[0] / smaples_num
            counts = np.sum(X_class, axis=0)
            self.word_probs[i, :] = (counts + 1) / (np.sum(counts) + features_num)

    
    # I used logs for calculating probs to prevent underflowing (n_i is the number of x_i word occurunce)
    # log(P(C)*product(P(x_i|C)^n_i)) = log(P(C)) + sum(n_i*log(P(x_i|C)))
    def get_log_probs(self, X):
        smaples_num = X.shape[0]
        classes_num = len(self.classes)
        log_probs = np.zeros((smaples_num, classes_num))
        for i in range(classes_num):
            # log(P(C_k))
            class_prob = np.log(self.class_probs[i])
            # log(P(X | C_k)) = sum over features of n_i * log(P(x_i | C_k))
            sum_word_probs = np.dot(X, np.log(self.word_probs[i, :]))
            log_probs[:, i] = class_prob + sum_word_probs  
        return log_probs
    
    def predict(self, X):
        log_probs = self.get_log_probs(X)
        return np.argmax(log_probs, axis=1)
```
